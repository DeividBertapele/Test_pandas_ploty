```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = ['2016','2017','2018','2019','2020','2021']
y = [2,15,1,10,5,12]

plt.rcParams['font.size'] = 14
plt.figure(figsize=(12, 5))
plt.plot(x,y, label = 'Primeiro Grfico', color='#2561BE', ls = '--', lw =2 , marker = 'd', ms = 10)
plt.title('Vendas x Ano', fontsize=18, color='#a83240')
plt.xlabel('Ano', fontsize=18, color='#a84e32')
plt.ylabel('Vendas', fontsize=18, color='#a84e32')
plt.xticks(rotation=45)
plt.legend()

plt.show()
```
